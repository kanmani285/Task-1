# ***3. Load the dataset***

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

# ***5. Checking for Missing Cells***

```python
df.isnull().sum()
```

```python
null_percent = (df.isnull().sum() / len(df)) * 100
null_percent[null_percent > 0]
```

# ***6.Handling Missing Files***

```python
# Fill 'country' , 'director' , 'duration' , 'cast' and 'rating' with 'Unknown', drop rows where title or type is missing
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['director'].fillna('Unknown', inplace= True)
df.dropna(subset=['title', 'type'], inplace=True)
```

# ***7. Remove Duplicates***

```python
df.drop_duplicates(inplace=True)
```

# ***8. Standardizing Text Columns***

```python
df['type'] = df['type'].str.strip().str.title()
df['rating'] = df['rating'].str.strip().str.upper()
```

# ***9. Convert Date columns***

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

# ***10. Clean duration Column***

```python

# Fill NA and split duration
df['duration'].fillna('0 min', inplace=True)
df[['duration_int', 'duration_type']] = df['duration'].str.extract(r'(\d+)\s*(\w+)')
# Convert 'duration_int' to numeric, handle errors, fill NaNs with 0, and then convert to int
df['duration_int'] = pd.to_numeric(df['duration_int'], errors='coerce').fillna(0).astype(int)
```

# ***11. Exploratory Analysis***

## Step 2: Exploratory Data Analysis

Let’s create visualizations and explore patterns in the data.

***Movie Vs TV Show***

```python
sns.countplot(data=df, x='type')
plt.title("Distribution of Content Type")
plt.show()
```

***Top 10 Countries***

```python
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='barh', color='skyblue')
plt.title("Top 10 Countries by Content")
plt.xlabel("Number of Titles")
plt.gca().invert_yaxis()
plt.show()
```

***Ratings***

```python
df['rating'].value_counts().head(10).plot(kind='bar', color='orange')
plt.title("Top 10 Content Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

***Content Over Years***

```python
df['release_year'] = df['release_year'].fillna(0).astype(int)
df[df['release_year'] > 1950]['release_year'].value_counts().sort_index().plot(figsize=(14, 5))
plt.title("Content Released Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Releases")
plt.show()
```

***Content Added Over Months***

```python
df['month_added'] = df['month_added'].fillna(0).astype(int)
df['month_added'].replace(0, np.nan, inplace=True)

month_counts = df['month_added'].value_counts().sort_index()
month_counts.index = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                      'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

sns.barplot(x=month_counts.index, y=month_counts.values, palette='Blues')
plt.title("Content Added by Month")
plt.ylabel("Number of Titles")
plt.xlabel("Month")
plt.show()
```

***Word Cloud Of Titles***

```python
from wordcloud import WordCloud

title_text = ' '.join(df['title'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='black').generate(title_text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Most Common Words in Titles")
plt.show()
```
